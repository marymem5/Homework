```python
import pandas as pd
import numpy as np
```

```python
city_students = "city_students.csv "
city_schools = "city_schools.csv"
```

```python
student_data = pd.read_csv(city_students)
school_data = pd.read_csv(city_schools)
```

```python
#combine csv files into one dataframe
combined_school_student_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#find number of schools in district
school_count=len(combined_school_student_data["school_name"].unique())
```

```python
#find number of students in district
student_count=combined_school_student_data["student_name"].count()
```

```python
#identify the individual values for school budgets and sum
school_budgets=combined_school_student_data.budget.unique()
total_budget=school_budgets.sum()
```

```python
#find average math score
mean_math=combined_school_student_data["math_score"].mean()
```

```python
#find average reading score
mean_reading=combined_school_student_data["reading_score"].mean()
```

```python
#find overall pass rate
overall_pass_rate=(mean_math + mean_reading)/2
```

```python
#find number of students that passed math
mathpass_count_df=combined_school_student_data.loc[combined_school_student_data["math_score"]>69,["student_name","math_score"]]
mathpass_student_count=mathpass_count_df["student_name"].count()
```

```python
#calculate percent of students that passed math
percent_passmath=mathpass_student_count/student_count
```

```python
#find number then percent of students that passed reading
reading_pass_df=combined_school_student_data.loc[combined_school_student_data["reading_score"]>69,["student_name","reading_score"]]
readingpass_student_count=reading_pass_df["student_name"].count()
percent_passreading=readingpass_student_count/student_count
```

```python
#create the Summary Table
District_Summary=pd.DataFrame({"Total Schools":[school_count],"Total Students":[student_count],"Total Budget":[total_budget],"Average Math Score":[mean_math],"Average Reading Score":[mean_reading],"Overall Pass Rate":[overall_pass_rate]," Percent Students Passing Math":[percent_passmath],"Percent Students Passing Reading":[percent_passreading]})
District_Summary
```

```python
#set variables for school summaries
group_school=combined_school_student_data.groupby(["school_name"])
student_count2=group_school.size()
school_budget = group_school["budget"].first()
school_type = group_school["type"].first()
budget_per_student=school_budget/student_count2
school_avg_math = group_school["math_score"].mean()
school_avg_reading= group_school["reading_score"].mean()
school_pass_math=combined_school_student_data[combined_school_student_data["math_score"]>69].groupby(["school_name"]).size()
school_percent_passmath=school_pass_math/student_count2
school_pass_reading=combined_school_student_data[combined_school_student_data["reading_score"]>69].groupby(["school_name"]).size()
school_percent_passreading=school_pass_reading/student_count2
school_overall_pass=(school_percent_passmath + school_percent_passreading)/2
```

```python
#name columns for new variables
schoolstats={"Total Students":student_count2, "Budget": school_budget, "Type": school_type, "Budget per Student":budget_per_student, "Average Math Score":school_avg_math,
            "Average Reading Score":school_avg_reading, "Percent Passing Math":school_percent_passmath, "Percent Passing Reading": school_percent_passreading,
            "Overall Pass Rate":school_overall_pass}
```

```python
#pass the variables into a dataframe
School_Summary=pd.DataFrame(schoolstats)
#print dataframe of school stats for 2nd part of exercise.
School_Summary
```

```python
#Show top schools by sorting descending overal pass rate
School_Summary.sort_values(["Overall Pass Rate"], ascending=False).head()
```

```python
#sort schools by descending overall pass rate to show lowest 
School_Summary.sort_values(["Overall Pass Rate"], ascending=True).head()
```

```python
#group the data by school for each grade level and display the average math scores: https://chrisalbon.com/python/data_wrangling/pandas_apply_operations_to_groups/
math_9th_grade=combined_school_student_data[combined_school_student_data["grade"]=="9th"].groupby("school_name")["math_score"].mean()
math_10th_grade=combined_school_student_data[combined_school_student_data["grade"]=="10th"].groupby("school_name")["math_score"].mean()
math_11th_grade=combined_school_student_data[combined_school_student_data["grade"]=="11th"].groupby("school_name")["math_score"].mean()
math_12th_grade=combined_school_student_data[combined_school_student_data["grade"]=="12th"].groupby("school_name")["math_score"].mean()
grouped_math_scores={"9th":math_9th_grade, "10th":math_10th_grade, "11th":math_11th_grade, "12th":math_12th_grade}
Math_Scores_by_Grade=pd.DataFrame(grouped_math_scores)
Math_Scores_by_Grade
```

```python
#re-run for reading
reading_9th_grade=combined_school_student_data[combined_school_student_data["grade"]=="9th"].groupby("school_name")["reading_score"].mean()
reading_10th_grade=combined_school_student_data[combined_school_student_data["grade"]=="10th"].groupby("school_name")["reading_score"].mean()
reading_11th_grade=combined_school_student_data[combined_school_student_data["grade"]=="11th"].groupby("school_name")["reading_score"].mean()
reading_12th_grade=combined_school_student_data[combined_school_student_data["grade"]=="12th"].groupby("school_name")["reading_score"].mean()
grouped_reading_scores={"9th":reading_9th_grade, "10th":reading_10th_grade, "11th":reading_11th_grade, "12th":reading_12th_grade}
Reading_Scores_by_Grade=pd.DataFrame(grouped_reading_scores)
Reading_Scores_by_Grade
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["< 585", "585-614", "615-644", "645-675"]
```

```python
#pandas ins binning activity
School_Summary["Spending Per Student"] = pd.cut(School_Summary["Budget per Student"], spending_bins, labels=group_names)
```

```python
#reduce columns - pandas imdb activity
Reduced_School_Summary = School_Summary[["Spending Per Student","Average Math Score","Average Reading Score","Percent Passing Math","Percent Passing Reading","Overall Pass Rate"]]
```

```python
#Group by the bins - pandas 3, activity 3
Reduced_School_Summary=Reduced_School_Summary.groupby("Spending Per Student")
Reduced_School_Summary.max()
```

```python
pass_mathtype=combined_school_student_data[combined_school_student_data["math_score"]>69].groupby("type").size()
pass_readingtype=combined_school_student_data[combined_school_student_data["reading_score"]>69].groupby("type").size()

type_table=combined_school_student_data[["type","reading_score","math_score"]]
type_data=type_table.groupby(["type"])
type_studentcount=type_data["math_score"].count()
```

```python
average_math=type_data["math_score"].mean()
average_reading=type_data["reading_score"].mean()

percentpassmath=pass_mathtype/type_studentcount
percentpassreading=pass_readingtype/type_studentcount
Overallpasstype=(percentpassmath+percentpassreading)/2
```

```python
typestats={"Average Math Score":average_math, "Average Reading Score":average_reading, "Percent Passing Math":percentpassmath, "Percent Passing Reading":percentpassreading, "Overall Pass Rate": Overallpasstype}
```
